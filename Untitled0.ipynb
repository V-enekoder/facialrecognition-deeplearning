```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt # Para visualizar fotos y resultados
from google.colab import drive
drive.mount('/content/drive')
```

```python
#Preprocesamiento
import cv2
import os
from google.colab import drive

# 1. Asegurar que el Drive está montado
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

# 2. Configuración de rutas
base_path = '/content/drive/MyDrive/dataset-DL'
detector_rostros = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# 3. Obtener solo las carpetas de Victor (excluyendo las que ya digan 'recortado')
carpetas_victor = [f for f in os.listdir(base_path)
                   if f.startswith('victor_') and not f.endswith('_recortado')]

print(f"Carpetas encontradas para procesar: {carpetas_victor}")

for carpeta in carpetas_victor:
    ruta_entrada = os.path.join(base_path, carpeta)
    ruta_salida = os.path.join(base_path, f"{carpeta}_recortado")

    # Crear la carpeta de salida si no existe
    if not os.path.exists(ruta_salida):
        os.makedirs(ruta_salida)
        print(f"--- Procesando: {carpeta} -> {carpeta}_recortado ---")

    contador = 0
    archivos = os.listdir(ruta_entrada)

    for archivo in archivos:
        img_path = os.path.join(ruta_entrada, archivo)
        img = cv2.imread(img_path)

        if img is None:
            continue

        # Convertir a gris para el detector
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        # Detectar rostros
        rostros = detector_rostros.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

        # Si encuentra al menos un rostro, recortamos el primero
        if len(rostros) > 0:
            (x, y, w, h) = rostros[0]
            # Añadimos un pequeño margen para que no quede tan pegado
            margen = int(w * 0.1)
            y_start = max(0, y - margen)
            y_end = min(img.shape[0], y + h + margen)
            x_start = max(0, x - margen)
            x_end = min(img.shape[1], x + w + margen)

            cara_recortada = img[y_start:y_end, x_start:x_end]

            # Guardar el recorte en la nueva carpeta
            cv2.imwrite(os.path.join(ruta_salida, f"crop_{archivo}"), cara_recortada)
            contador += 1

    print(f"Finalizado: {contador} rostros guardados en {ruta_salida}")

print("\n✅ ¡Proceso completado con éxito!")
```

```python
import numpy as np

# Configuración de dimensiones
img_height, img_width = 150, 150
batch_size = 32

# 1. Cargar el dataset de entrenamiento (70%)
train_ds = tf.keras.utils.image_dataset_from_directory(
  data_path,
  validation_split=0.3, # Aquí definimos el 30% para validación
  subset="training",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size,
  label_mode='categorical' # Importante para clasificación multiclase
)

# 2. Cargar el dataset de validación (30%)
val_ds = tf.keras.utils.image_dataset_from_directory(
  data_path,
  validation_split=0.3,
  subset="validation",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size,
  label_mode='categorical'
)

# Obtener los nombres de las clases (carpetas) para verificar
class_names = train_ds.class_names
print(f"\nClases detectadas: {class_names}")

# Optimización para que el entrenamiento sea más rápido
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
# Capas de aumento de datos
data_augmentation = keras.Sequential([
  layers.RandomFlip("horizontal", input_shape=(img_height, img_width, 3)),
  layers.RandomRotation(0.1),
  layers.RandomZoom(0.1),
])

# Crear la Red Neuronal Convolucional (CNN)
model = models.Sequential([
  data_augmentation,
  layers.Rescaling(1./255), # Normalizar píxeles de 0-255 a 0-1

  layers.Conv2D(32, (3, 3), activation='relu'),
  layers.MaxPooling2D(2, 2),

  layers.Conv2D(64, (3, 3), activation='relu'),
  layers.MaxPooling2D(2, 2),

  layers.Conv2D(128, (3, 3), activation='relu'),
  layers.MaxPooling2D(2, 2),

  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dropout(0.5), # Evita que el modelo "memorice" (Overfitting)
  layers.Dense(len(class_names), activation='softmax') # Una neurona por cada clase
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
import tensorflow as tf

# 2. Configurar Early Stopping (El "freno" automático)
callback = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=15,
    restore_best_weights=True
)

# 3. Entrenar hasta el límite de su capacidad
print("--- Iniciando entrenamiento inteligente del Agente ---")
history = model.fit(
  train_ds,
  validation_data=val_ds,
  epochs=200,          # Ponemos 200 como límite, pero el callback lo detendrá antes
  callbacks=[callback] # Aplicamos el freno automático
)

print("\n✅ Entrenamiento finalizado en el punto óptimo.")
```

```python
from google.colab import files
from tensorflow.keras.utils import load_img, img_to_array

uploaded = files.upload()

for fn in uploaded.keys():
  # Cargar la imagen subida
  path = fn
  img = load_img(path, target_size=(img_height, img_width))
  img_array = img_to_array(img)
  img_array = tf.expand_dims(img_array, 0) # Crear un batch

  predictions = model.predict(img_array)
  score = tf.nn.softmax(predictions[0])

  print(
    "Esta imagen pertenece a {} con un {:.2f} por ciento de confianza."
    .format(class_names[np.argmax(predictions)], 100 * np.max(predictions))
  )
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import numpy as np

# 1. Obtener etiquetas reales y predicciones
y_true = []
y_pred = []

for x, y in val_ds:
    y_true.extend(np.argmax(y, axis=1))
    preds = model.predict(x)
    y_pred.extend(np.argmax(preds, axis=1))

# 2. Crear la matriz
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicción (IA)')
plt.ylabel('Realidad (Tú)')
plt.title('Matriz de Confusión')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Crear una figura con dos subgráficos
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(len(acc))

plt.figure(figsize=(12, 5))

# Gráfica de Precisión (Accuracy)
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Entrenamiento')
plt.plot(epochs_range, val_acc, label='Validación')
plt.title('Precisión del Modelo')
plt.legend(loc='lower right')
plt.grid(True)

# Gráfica de Pérdida (Loss)
plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Entrenamiento')
plt.plot(epochs_range, val_loss, label='Validación')
plt.title('Valor de Pérdida')
plt.legend(loc='upper right')
plt.grid(True)

plt.show()
```
